5. The Modulo Architect 

```python
seconds = input ("Enter seconds: ")
print (seconds)
try:
    seconds_int = int(seconds)
except ValueError:
    print("Please enter a number.")
    exit()
houres =  seconds_int // 3600
print (houres)
remain_second = seconds_int % 3600
print (remain_second)
min = remain_second // 60
print(min)
sec = remain_second % 60
print(houres, ":", min, ":", sec)
```
